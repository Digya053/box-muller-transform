```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
# draw random samples from the uniform distribution
u1 = np.random.rand(1000)
u2 = np.random.rand(1000)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Uniformly distributed samples')
ax1.hist(u1)
ax1.set(xlabel='u1')
ax2.hist(u2)
ax2.set(xlabel='u2')
```

### Generate normal distribution

```python
r = np.sqrt(-2*np.log(u1))

theta = 2*np.pi*u2

x = r*np.cos(theta);
y = r*np.sin(theta);
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Normal Distribution')
ax1.hist(x)
ax1.set(xlabel='x')
ax2.hist(y)
ax2.set(xlabel='y')
```
